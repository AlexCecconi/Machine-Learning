# Classificação Linear

Com esta equação nos encontramos os pesos \
$\overrightarrow{w} =  (  X^{t}X )^{-1}  X^{t}\overrightarrow{y}   $

```python
#y = np.array([1,2,3,4])
y = y_train.values
```

```python
#X = np.array([[1,2,7,5],     [2,3,9,1],     [2,4,3,1],     [5,6,1,1]])
#Estou usando a base de treino, transformo o df em matrix e depois converto string em float.
X = X_train.values.astype(float)
```

```python
X_T = X.transpose()
#X_T
```

```python
X_Mult = X_T.dot(X)
#X_Mult
```

```python
df_inv = np.linalg.inv(X_Mult)
#df_inv
```

```python
pseudo_inv = df_inv.dot(X_T)
```

```python
omega1 = pseudo_inv.dot(y)
#omega1
```

Abaixo temos o classificador, podemos definir um limiar em que teremos as classes\
$h(\overrightarrow{w}) = \overrightarrow{w} \cdot \overrightarrow{x}$

```python
#Aqui estou aplicando os omegas treinados na base de teste para tentar predizer estes caras.
F_x1 = omega1*X_test.values.astype(float)
#F_x1
```

```python
H_x1 = np.where( (F_x1).sum(axis=1) >1.5,1,0)
H_x1
```

```python
Acuracia = H_x1.sum() / H_x1.shape[0] 
round(Acuracia,3)
```

Aqui temos os betas com fator de regularização\
$\overrightarrow{w} =  (  \gamma I + X^{t}X )^{-1}  X^{t}\overrightarrow{y}   $

```python
gama = 150
omega2 = np.linalg.inv( np.multiply(gama,np.identity(X_train.iloc[0,:].shape[0])) + X.T.dot(X)).dot(X.T).dot(y)
omega2
```

```python
F_x2 = omega2*X_test.values.astype(float)
H_x2 = np.where( (F_x2).sum(axis=1) >1.5,1,0)
```

```python
Acuracia_Reg = H_x2.sum() / H_x2.shape[0] 
Acuracia_Reg
```

```python
Valores = []
gama = 0
for gama in range(0,1000,10):
    omega2 = np.linalg.inv( np.multiply(gama,np.identity(X_train.iloc[0,:].shape[0])) + X.T.dot(X)).dot(X.T).dot(y)
    F_x2 = omega2*X_test.values.astype(float)
    H_x2 = np.where( (F_x2).sum(axis=1) >1.5,1,0)
    Acuracia_Reg = H_x2.sum() / H_x2.shape[0] 
    Valores.append(Acuracia_Reg)        
            
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(0,1000,10),Valores)
plt.show()
```
